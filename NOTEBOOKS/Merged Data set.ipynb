```python
import pandas as pd
```

```python
customers=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTEgwFZoArQxzrq0jWWDQVBppTGYUV_Ka0N0jbDG5WIS-rIOv6SRD-ew93imnCK3ExrGBuS5J7lhb5f/pub?output=csv")
orders=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSt3FK3N-slM6ExBo-Wz2NSHwXAmxwvUCTHelAIsUkylUbtq3RTUpWIcKPfHS7ZYdzmEq13z4DnYslm/pub?output=csv")
order_items = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRVK_VN5jeWgh7XnIQVB5QuObbuVP_Z3gVjS5rpt_oj6eyDrS4itg0LHiOJRYH1JMDVHqwM1eeSN4XX/pub?output=csv")
products=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQ9acK5QmnoV3JSiwEpd1smjx99OmCP-XSNFBqUaiahkxf2C3GH4eJQ6n1KMf4Anv2ti8LjoIYPjv0Z/pub?output=csv")
payments=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQlbhQCuXYjxFmIeTSeTFXhGQMGe5f9eP-svg8-ucyrfpoEQC9rt-drjoqkLulkdp3jMAjDwvOOWYol/pub?output=csv")
reviews=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQDgZ9y9wIfUpD7jZqd0Hz0Oe5cePVzPljTNZUP0PZEysakMjvUugbJef3bOyX_6DwGpzkOqvWW8Cp8/pub?output=csv")
sellers=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSjhMFqKZY_YcZqvsgEzqNxgxB7P1UlZi1hIa8E2t3VUa7ueVyNWHDx7EDv5AXwq_bNxuNOi14cwC5p/pub?output=csv")
translations=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSN6bxQKqtTzRrVUx2sojqirjBicTYuvh3UaVJBuVAi7fyrj_NtMQ97jPwaGcVn3NVokjQIPsCjal8u/pub?output=csv")
geolocation=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTstVWkA8eB8RqVhHABsJ_NpMMvK9B903qZJEZmgny9TgJvWfk9GqOS6Z0ZBEQVOU_DhAQzXdNU9T7Q/pub?output=csv")
```

```python
# Merge customers with orders
customer_orders = pd.merge(orders, customers, on="customer_id", how="inner")
```

```python
# Merge with order_items
combined_data_set = pd.merge(customer_orders, order_items, on="order_id", how="inner")
```

```python
# Merge with products
products.rename(columns={'Unnamed: 0': 'product_id'}, inplace=True)
combined_data_set = pd.merge(combined_data_set, products, on="product_id", how="left")
```

```python
# Merge with sellers
combined_data_set = pd.merge(combined_data_set, sellers, on="seller_id", how="left")
```

```python
# Optional: Merge with payments (for spend insights)
payments.rename(columns={'Unnamed: 0': 'order_id'}, inplace=True)
combined_data_set = pd.merge(combined_data_set, payments, on="order_id", how="left")
```

```python
# Optional: Merge with reviews (for sentiment)
combined_data_set = pd.merge(combined_data_set, reviews[['order_id', 'review_score']], on="order_id", how="left")
```

```python
# Optional: Translate product category
combined_data_set = pd.merge(combined_data_set, translations, on="product_category_name", how="left")
```

```python
# Rename columns for clarity
combined_data_set.rename(columns={
    'product_category_name_english': 'product_category',
    'customer_state': 'state'
}, inplace=True)
```

```python

# Quick check on the merged result
print("Final shape of combined_data_set:", combined_data_set.shape)
combined_data_set.head()
```

```python
# Save to the same folder as your notebook
combined_data_set.to_csv("combined_data_set.csv", index=False)
```
